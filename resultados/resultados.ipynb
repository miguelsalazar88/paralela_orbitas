```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tamanios = [50000,100000,500000,1000000,5000000,10000000]
yCython = []
yPython = []
```

```python
paths = [
    '/home/miguel/Desktop/computacion_paralela/ejercicio_orbitas/resultados/int-size-50000.csv',
    '/home/miguel/Desktop/computacion_paralela/ejercicio_orbitas/resultados/int-size-100000.csv',
    '/home/miguel/Desktop/computacion_paralela/ejercicio_orbitas/resultados/int-size-500000.csv',
    '/home/miguel/Desktop/computacion_paralela/ejercicio_orbitas/resultados/int-size-1000000.csv',
    '/home/miguel/Desktop/computacion_paralela/ejercicio_orbitas/resultados/int-size-5000000.csv',
    '/home/miguel/Desktop/computacion_paralela/ejercicio_orbitas/resultados/int-size-10000000.csv'
]
```

```python
for path in paths:
    df = pd.read_csv(path)
    yCython.append(df['Cython'].mean())
    yPython.append(df['Python'].mean())
```

```python
plt.plot(tamanios, yPython, label='Python')
plt.plot(tamanios, yCython, label='Cython')
plt.xlabel('steps')
plt.ylabel('tiempo (s)')
plt.legend()
```

```python
res = []
for i in range(len(yPython)):
    veces = yPython[i] / yCython[i]
    res.append(veces)
    print(f'Cython es {veces} veces mas rapido que Python')

resultado = sum(res)/len(yCython)
print(resultado)
11.98
```

```python

fig, ax = plt.subplots()
x = np.arange(len(tamanios))
y1 = yCython
y2 = yPython
width = 0.2
plt.figure(figsize=(15,10))
# plot data in grouped manner of bar type
plt.bar(x, y1, width)
plt.bar(x-0.2, y2, width)
plt.xticks(x, tamanios)
plt.title("Cython vs. Python: Performance")
plt.xlabel("Steps (N)")
plt.ylabel("Segundos (s)")
plt.legend(['Python', 'Cython'])


plt.show()
```
